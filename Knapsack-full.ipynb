```python
from collections import OrderedDict

NUMBER_OF_OBJECTS = int(input())

KNAPSACK_VOLUME_LIMIT = int(input())
KNAPSACK_WEIGHT_LIMIT = int(input())

item_FEAT = []

for i in range(NUMBER_OF_OBJECTS):
    a = list(map(int, input().split()))
    item_FEAT.append(a)
    
transpose = [*zip(*item_FEAT)]

value = transpose[0]
volume = transpose[1]
weight = transpose[2]

aDict = {}


def greedy_value():
    ans1 = []
    assert len(volume) == len(value) == len(weight)
    n_items = len(volume)

    sorted_indices = sorted(range(n_items), key=lambda i: value[i], reverse = True)

    KNAPSACK_VOLUME_REMAINING = KNAPSACK_VOLUME_LIMIT
    KNAPSACK_WEIGHT_REMAINING = KNAPSACK_WEIGHT_LIMIT

    gained_value = 0
    for i in sorted_indices:
        if volume[i] <= KNAPSACK_VOLUME_REMAINING and weight[i] <= KNAPSACK_WEIGHT_REMAINING:

            KNAPSACK_WEIGHT_REMAINING -= weight[i]
            KNAPSACK_VOLUME_REMAINING -= volume[i]
            gained_value += value[i]

            ans1.append(i)
    aDict[gained_value] = ans1


def greedy_weight():

    ans2 = []

    assert len(volume) == len(value) == len(weight)
    n_items = len(weight)

    sorted_indices = sorted(range(n_items), key=lambda i: weight[i], reverse = False)

    KNAPSACK_VOLUME_REMAINING = KNAPSACK_VOLUME_LIMIT
    KNAPSACK_WEIGHT_REMAINING = KNAPSACK_WEIGHT_LIMIT

    gained_value = 0
    for i in sorted_indices:
        if volume[i] <= KNAPSACK_VOLUME_REMAINING and weight[i] <= KNAPSACK_WEIGHT_REMAINING:

            KNAPSACK_WEIGHT_REMAINING -= weight[i]
            KNAPSACK_VOLUME_REMAINING -= volume[i]
            gained_value += value[i]

            ans2.append(i)
    aDict[gained_value] = ans2

def greedy_volume():

    ans3 = []

    assert len(volume) == len(value) == len(weight)
    n_items = len(volume)

    sorted_indices = sorted(range(n_items), key=lambda i: volume[i], reverse = False)

    KNAPSACK_VOLUME_REMAINING = KNAPSACK_VOLUME_LIMIT
    KNAPSACK_WEIGHT_REMAINING = KNAPSACK_WEIGHT_LIMIT

    gained_value = 0
    for i in sorted_indices:
        if volume[i] <= KNAPSACK_VOLUME_REMAINING and weight[i] <= KNAPSACK_WEIGHT_REMAINING:

            KNAPSACK_WEIGHT_REMAINING -= weight[i]
            KNAPSACK_VOLUME_REMAINING -= volume[i]
            gained_value += value[i]

            ans3.append(i)
    aDict[gained_value] = ans3


def greedy_value_per_volume():

    ans4 = []

    assert len(volume) == len(value) == len(weight)
    n_items = len(volume)

    sorted_indices = sorted(range(n_items), key=lambda i: value[i] / volume[i], reverse = True)

    KNAPSACK_VOLUME_REMAINING = KNAPSACK_VOLUME_LIMIT
    KNAPSACK_WEIGHT_REMAINING = KNAPSACK_WEIGHT_LIMIT

    gained_value = 0
    for i in sorted_indices:
        if volume[i] <= KNAPSACK_VOLUME_REMAINING and weight[i] <= KNAPSACK_WEIGHT_REMAINING:

            KNAPSACK_WEIGHT_REMAINING -= weight[i]
            KNAPSACK_VOLUME_REMAINING -= volume[i]
            gained_value += value[i]

            ans4.append(i)
    aDict[gained_value] = ans4


def greedy_value_per_weight():

    ans5 = []

    assert len(volume) == len(value) == len(weight)
    n_items = len(volume)

    sorted_indices = sorted(range(n_items), key=lambda i: value[i] / weight[i], reverse = True)

    KNAPSACK_VOLUME_REMAINING = KNAPSACK_VOLUME_LIMIT
    KNAPSACK_WEIGHT_REMAINING = KNAPSACK_WEIGHT_LIMIT

    gained_value = 0
    for i in sorted_indices:
        if volume[i] <= KNAPSACK_VOLUME_REMAINING and weight[i] <= KNAPSACK_WEIGHT_REMAINING:

            KNAPSACK_WEIGHT_REMAINING -= weight[i]
            KNAPSACK_VOLUME_REMAINING -= volume[i]
            gained_value += value[i]

            ans5.append(i)
    aDict[gained_value] = ans5


greedy_value()
greedy_weight()
greedy_volume()
greedy_value_per_volume()
greedy_value_per_weight()

dDict = OrderedDict(sorted(aDict.items(), 
                              key=lambda kv: kv[0], reverse=True))

print(list(dDict.keys())[0])

for elem in list(dDict.values())[0]:
    print(elem, end = ' ')

```
